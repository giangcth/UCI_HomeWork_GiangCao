```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete
```

```python
# total budget

school_budget = school_data_complete["budget"].unique()

totalbudget = school_budget.sum()

totalbudget
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_count = len(school_data_complete["school_name"].unique())

student_count = school_data_complete["Student ID"].count()
average_Mscore = school_data_complete["math_score"].mean()
average_Rscore = school_data_complete["reading_score"].mean()

total_Mpass = school_data_complete.loc[(school_data_complete["math_score"] >= 70)]
percentMpass = total_Mpass["math_score"].count()/student_count*100

total_Rpass = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]
percentRpass = total_Rpass["reading_score"].count()/student_count*100

RMpass = pd.merge(total_Mpass,total_Rpass, on="Student ID", how="inner")
totalpassRM = RMpass["Student ID"].count()
percent_totapassRM = totalpassRM/student_count*100

district_Summary = pd.DataFrame({"Total Schools": [school_count], "Total Students": student_count, "Total Budget": totalbudget,
                              "Average Math Score": average_Mscore, "Average Reading Score": average_Rscore,
                              "%Passing Math": percentMpass, "%Passing Reading": percentRpass, 
                               "%Overal Passing": percent_totapassRM })

district_Summary["Total Students"]=district_Summary["Total Students"].map("{:,}".format)
district_Summary["Total Budget"]=district_Summary["Total Budget"].map("${:,.2f}".format)
district_Summary["%Passing Math"]=district_Summary["%Passing Math"].map("{:.2f}%".format)
district_Summary["%Passing Reading"]=district_Summary["%Passing Reading"].map("{:.2f}%".format)
district_Summary["%Overal Passing"]=district_Summary["%Overal Passing"].map("{:.2f}%".format)

district_Summary
```

## School Summary

```python
school_data=school_data_complete.groupby(["school_name"])

totalstudent_perschool = school_data["Student ID"].count()
totalbudget_perschool = school_data["budget"].unique().astype(int)
budget_perstudent = totalbudget_perschool/totalstudent_perschool
averageMath_perschool = school_data["math_score"].mean()
averageReading_perschool = school_data["reading_score"].mean()
                                               
totalstudent_perschool
```

```python
totalstudents_passing_math = school_data_complete[(school_data_complete["math_score"]>=70)]
students_perschool_mathpass = totalstudents_passing_math.groupby(["school_name"]).count()["Student ID"]
percentage_perschools_mathpass = (students_perschool_mathpass)/(totalstudent_perschool)*100

percentage_perschools_mathpass
```

```python
totalstudents_passing_reading = school_data_complete[(school_data_complete["reading_score"]>=70)]
students_perschool_readingpass = totalstudents_passing_reading.groupby(["school_name"]).count()["Student ID"]
percentage_perschools_readingpass = (students_perschool_readingpass/totalstudent_perschool)*100

percentage_perschools_readingpass
```

```python
totalstudents_passing_mathreading = school_data_complete[(school_data_complete["math_score"]>=70) & 
                                                         (school_data_complete["reading_score"]>=70)]
students_perschool_passing_mathreading = totalstudents_passing_mathreading.groupby(["school_name"]).count()["Student ID"]
percentage_perschools_passing_mathreading = (students_perschool_passing_mathreading/totalstudent_perschool)*100

percentage_perschools_passing_mathreading 
```

```python
School_Summary = pd.DataFrame({"Total Students": totalstudent_perschool, "Total School Budget": totalbudget_perschool,
                              "Per Student Budget": budget_perstudent, "Average Math Score": averageMath_perschool,
                              "Average Reading Score": averageReading_perschool, 
                               "% Passing Math": percentage_perschools_mathpass, 
                              "% Passing Reading": percentage_perschools_readingpass,
                               "% Overall Passing": percentage_perschools_passing_mathreading})

School_Summary["Total Students"]=School_Summary["Total Students"].map("{:,}".format)
School_Summary["Total School Budget"]=School_Summary["Total School Budget"].map("${:,.2f}".format)
School_Summary["Per Student Budget"]=School_Summary["Per Student Budget"].map("${:,.2f}".format)
School_Summary["Average Math Score"]=School_Summary["Average Math Score"].map("{:,.2f}".format)
School_Summary["Average Reading Score"]=School_Summary["Average Reading Score"].map("{:,.2f}".format)
School_Summary["% Passing Math"]=School_Summary["% Passing Math"].map("{:.2f}%".format)
School_Summary["% Passing Reading"]=School_Summary["% Passing Reading"].map("{:.2f}%".format)
School_Summary["% Overall Passing"]=School_Summary["% Overall Passing"].map("{:.2f}%".format)

School_Summary
```
